```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,how='left')
```

```python
final_df.head(2)
```

```python
final_df.Country.value_counts()
```

```python
final_df.Country.value_counts().index
```

```python
country_names = final_df.Country.value_counts().index[:3]
```

```python
country_val = final_df.Country.value_counts().values[:3]
```

```python
# Slightly separate each slice
explode = [0.05] * len(country_names)

# Use a vibrant colormap
colors = plt.cm.tab20.colors[:len(country_names)]

# Create pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    country_val,
    labels=country_names,
    autopct='%1.1f%%',
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=90,
    textprops={'fontsize': 12, 'color': 'black'},
    wedgeprops={'edgecolor': 'black'}
)

# Add title
plt.title("Country-wise Distribution", fontsize=16)

# Ensure everything fits well
plt.tight_layout()
plt.show()
```
